```python
from sklearn import datasets
iris = datasets.load_iris()
```

```python
iris
```

```python
iris.keys()
```

```python
import pandas as pd
x = pd.DataFrame(iris['data'], columns=iris['feature_names'])
```

```python
x
```

```python
y = pd.DataFrame(iris['target'], columns=['target_name'])
```

```python
y
```

```python
data = pd.concat([x,y],axis=1) #axis=1 為水平合併, 2 為垂直合併
```

```python
data
```

```python
import matplotlib.pyplot as plt
data.plot.scatter(x='petal length (cm)', y='petal width (cm)', c='target_name')# 分布圖
                        # (x軸, y軸, 顏色灰階)
```

```python
color = {0:'r', 1:'g', 2:'b'}
data['color'] = data['target_name'].map(color)
```

```python
data.plot.scatter(x='petal length (cm)', y='petal width (cm)', c='color')
```
